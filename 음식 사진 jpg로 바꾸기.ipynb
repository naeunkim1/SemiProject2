```python
from tqdm import tqdm
import os
import glob
from PIL import Image
from PIL import ImageFile

for f in food16:
    train_list=os.listdir('d:/'+f+'/train/images/')
    test_list=os.listdir('d:/'+f+'/test/images/')
    for trfile in tqdm(train_list):
        if str(trfile[-4:])=='jpeg':
            img=Image.open('d:/'+f+'/train/images/'+trfile)
            img.save('d:/'+f+'/train/images/'+str(trfile[:-4])+'jpg')
            img.close()
            if os.path.isfile('d:/'+f+'/train/images/'+trfile):
                os.remove('d:/'+f+'/train/images/'+trfile)
        else:
            pass
    for tefile in tqdm(test_list):
        if str(tefile[-4:])=='jpeg':
            img=Image.open('d:/'+f+'/test/images/'+tefile)
            img.save('d:/'+f+'/test/images/'+str(tefile[:-4])+'jpg')
            img.close()

            if os.path.isfile('d:/'+f+'/test/images/'+tefile):
                os.remove('d:/'+f+'/test/images/'+tefile)

        else:
            pass

        
```
